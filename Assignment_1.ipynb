## 1. Introduction to `.loc` and `.iloc`

### `.loc`:
- **Definition:** `.loc` is a label-based data selection method. It is used to access a group of rows and columns by labels or a boolean array.
- **Syntax:** `df.loc[row_label, column_label]`
- **Use Cases:**
  - Selecting rows based on labels.
  - Selecting specific rows and columns by label.

### `.iloc`:
- **Definition:** `.iloc` is a position-based data selection method. It is used to access a group of rows and columns by integer positions (like a list in Python).
- **Syntax:** `df.iloc[row_index, column_index]`
- **Use Cases:**
  - Selecting rows based on their integer position.
  - Selecting specific rows and columns by position.

Let's explore these methods with examples.


### Select multiple rows by labels:


```python
# Select first and the fifth row
selected_rows = csv_data.loc[[0, 4]]
selected_rows
```

### Select specific rows and columns by labels:

```python
# Select first and fifth rows for latitude and longitude only
selected_data = csv_data.loc[[1, 4], ['latitude', 'longitude']]
selected_data
```

### Select specific rows by positions

```python
# Select the first and last rows
selected_rows_pos = csv_data.iloc[[0, -1]]
selected_rows_pos
```

### Select specific rows and columns by positions

```python
# Select the 'latitude' and 'total_rooms' columns for the second and third rows
selected_data_pos = csv_data.iloc[1:3, [1, 3]]
selected_data_pos
```

### Use iloc for slicing

```python
# Select the first three rows and the first two columns
sliced_data = csv_data.iloc[:3, :2]
sliced_data
```

```python
# What happens when you concatenate along different axes? Explain.
# Axis 0 (Rows): Concatenation along rows is straightforward, stacking DataFrames vertically. This is useful when combining datasets with the same columns but different rows (like appending data from different months).
# Axis 1 (Columns): Concatenation along columns aligns DataFrames side-by-side, which is useful when combining datasets with the same indices but different columns. If the indices don't align perfectly (different lengths), Pandas introduces NaN values.
```

# **Assignment 2: Merging**

```python
# Create two DataFrames with a common column and some non-overlapping data.
import pandas as pd

# DataFrame 1
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Sita', 'Ram', 'Gita', 'Hari'],
    'Age': [24, 30, 27, 35]
}
df1 = pd.DataFrame(data1)

# DataFrame 2
data2 = {
    'ID': [3, 4, 5, 6],
    'Name': ['Gita', 'Hari', 'Mina', 'Anita'],
    'Salary': [50000, 60000, 55000, 52000]
}
df2 = pd.DataFrame(data2)

df1, df2
```

```python
# Perform different types of merges (inner, left, right, outer) on these DataFrames.
inner_merge = pd.merge(df1, df2, on='ID', how='inner')

left_merge = pd.merge(df1, df2, on='ID', how='left')

right_merge = pd.merge(df1, df2, on='ID', how='right')

outer_merge = pd.merge(df1, df2, on='ID', how='outer')

# Display the results
print("Inner Merge:")
print(inner_merge)

print("\nLeft Merge:")
print(left_merge)

print("\nRight Merge:")
print(right_merge)

print("\nOuter Merge:")
print(outer_merge)

```

```python
# Creating a DataFrame
data = {
    'Category': ['A', 'B', 'A', 'C', 'B'],
    'Value': [10, 20, 15, 30, 25]
}
df = pd.DataFrame(data)


df
```

```python
# Use groupby() to Group Data and Calculate Aggregate Statistics
# Group by 'Category' and calculate aggregate statistics
grouped_stats = df.groupby('Category').agg({'Value': ['mean', 'sum', 'count']})

grouped_stats
```

```python
# Visualize the Results Using Bar Plots
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for mean values
plt.figure(figsize=(8, 4))
sns.barplot(x=grouped_stats.index, y=grouped_stats['Value']['mean'])
plt.title('Mean Value by Category')
plt.ylabel('Mean Value')
plt.xlabel('Category')
plt.show()

# Bar plot for sum values
plt.figure(figsize=(8, 4))
sns.barplot(x=grouped_stats.index, y=grouped_stats['Value']['sum'], palette='viridis')
plt.title('Sum Value by Category')
plt.ylabel('Sum Value')
plt.xlabel('Category')
plt.show()

# Bar plot for count values
plt.figure(figsize=(8, 4))
sns.barplot(x=grouped_stats.index, y=grouped_stats['Value']['count'], palette='inferno')
plt.title('Count by Category')
plt.ylabel('Count')
plt.xlabel('Category')
plt.show()
```
